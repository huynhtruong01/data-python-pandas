<h1 style="color:goldenrod">Learn Pandas</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./pokemon_data.csv')
```

```python
df.head(10)
```

## Get column, row, cell, header: .iloc()

```python
# get column
df['Name']
# get more column
df[['Name', 'Type 1', 'Attack']]

# .iloc()
# get number of row
df.head(10)
# get a row by index
df.iloc[1]
# get a row by slice
df.iloc[:7:]
# get list of index row
df.iloc[[0, 6, 12]]
# ------
df.iloc[1:3, 0:4] # .iloc[row, columns]
```

## Getting row by .loc()

```python
# getting row based on specific condition: .loc()
# it accept a array boolean of type Series. The rest are all the same .iloc()
df.loc[df['Name'] == 'Bulbasaur'] # by condition
df.loc[[1, 2, 156]]
df.loc[1]
df.loc[1:2]
```

## Filtering/Describing data

```python
# describing is generate descriptive statistics (tạo thống kê mô tả): .describe()
df.describe()
```

```python
#  filtering: sort_values()
df.sort_values(['Type 1', 'HP'], ascending=False, inplace=False)
```

## Making changes to the data

```python
# add column
df['Total'] = df['HP'] + df['Defense']
df.head(10)
```

```python
# remove column and update df
df = df.drop(columns='Total')
```

```python
df['Total'] = df.iloc[:, 4:12].sum(axis=1)
cols = [*df.columns.values]
total_rows = df.groupby('Type 2').sum()
total_rows
```

```python
max_total = total_rows['Total'].max()
max_total
```

```python
types1 = [type1 for type1, total in total_rows['Total'].items()]
```

```python
plt.bar(x=types1, height=total_rows['Total'])
plt.xticks(types1, rotation=90, size=8)
plt.xlabel('Types 1')
plt.ylabel('Total')
plt.show()
```
